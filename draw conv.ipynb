```python
import pandas as pd
import numpy as np
import h5py
import os
import sys
```

```python
f1 = h5py.File("a9a_Sophia_0.01_0.03_1.0_0.02_80u_256b_1_0.h5")
group = f1["rs_train_loss"]
a9a_done_losses = np.array(group)
a9a_opt_gap_done = a9a_done_losses - 0.3333472060757055
f1.close()
```

```python
# plot a9a_done_gap
import matplotlib.pyplot as plt
plt.plot(a9a_opt_gap_done)
plt.xlabel('iteration')
plt.ylabel('optimality gap')
plt.yscale('log')
plt.title('a9a optimality gap')
plt.show()
```
